### EMAISMAsimple moving average

 short-SMA > long-SMA ,    
 short-SMA < long-SMA ,    

```python
data['SMA_short'] = data['EMAI'].rolling(window=3).mean()
data['SMA_long'] = data['EMAI'].rolling(window=9).mean()
```

### position
 SMA_short > SMA_long ,  1   
 SMA_short < SMA_long ,  -1

```python
data['positions'] = np.where(data['SMA_short'] > data['SMA_long'],1,-1)
```
