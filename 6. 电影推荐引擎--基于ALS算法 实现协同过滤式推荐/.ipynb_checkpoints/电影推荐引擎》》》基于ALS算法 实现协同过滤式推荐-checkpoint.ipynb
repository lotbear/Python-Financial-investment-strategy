```python
#  Spark  Hadoop-yarn 
#  4 
print(' Spark ',str(sc.master))
```

```python
# 
#  sc.master  "local",
#  sc.master  "yarn""spark://master:7077"(Standalone)
#  HDFS 
global Path    
if sc.master[0:5]=="local" :
   Path="file:/home/lotbear/Big-Data/movie_rec/"
else:   
   Path="hdfs://master:9000/user/lotbear/data/movie_rec/"
```

```python
#  ratings  rawUserData
rawUserData = sc.textFile(Path+"ratings.csv")
print('',rawUserData.count())
```

```python
#  rawUserData  5 
for x in rawUserData.take(5): 
    print(x)
```

```python
#   
header = rawUserData.first()
rawUserData = rawUserData.filter(lambda x:x !=header)
rawUserData.take(5)
```

```python
#  map , rawUserData  3
#  RDD : rawRatings
rawRatings = rawUserData.map(lambda line: line.split(",")[:3] )
print('rawRatings ',type(rawRatings))
rawRatings.take(5)
```

####  ALS  Rating RDD:
Rating ( user , product , rating )    


```python
#  Rating 
from pyspark.mllib.recommendation import Rating
ratingsRDD = rawRatings.map(lambda x: (x[0],x[1],x[2]))
ratingsRDD.take(5)
```

```python
#  numRatings 
numRatings = ratingsRDD.count()
numRatings
```

```python
#  numUsers 
numUsers = ratingsRDD.map(lambda x: x[0] ).distinct().count()
numUsers 
```

```python
#  numMovies 
numMovies = ratingsRDD.map(lambda x: x[1]).distinct().count() 
numMovies
```

```python
# RDD 
ratingsRDD.persist()
```

####  ALS  

```python
from pyspark.mllib.recommendation import ALS
```

#### ALS.train()   
ALS.train ( ratings, rank, iterations=5, lambda_=0.01 )     
ALS.trainImplicit ( ratings, rank, iterations=5, lambda_=0.01 )      
rank iterations lambda    

 MatrixFactorizationModel   
rank:    
userFeatures:  X m \* rank )   
productFeatures:  Y rank \* n )   
 A  m \* n )  

```python
model = ALS.train(ratingsRDD, 10, 10, 0.01)
print(model)
```

###  >>> 

#### 
#### <font color=red>recommendProducts</font> 
 recommendProducts( user:Int, num:Int )   

 user  id  ,  num    

   

```python
model.recommendProducts(101,5)
```

```python
# 
# userID=100, movieID=179819,  rating 
model.predict(101, 179819)
```

#### 
#### <font color=red>recommendUsers</font>  

```python
model.recommendUsers(product=110,num=5)
```

```python
itemRDD = sc.textFile(Path+"movies.csv")
print('',itemRDD.count())
```

```python
#  rawUserData  5 
for x in itemRDD.take(5): 
    print(x)
```

```python
#   
header = itemRDD.first()
itemRDD = itemRDD.filter(lambda x:x !=header)
itemRDD.take(5)
```

```python
# id   
movieTitle= itemRDD.map( lambda line : line.split(","))\
                                   .map(lambda a: (float(a[0]),a[1]))\
                                   .collectAsMap()
print('movieTitle :',type(movieTitle))
print('',len(movieTitle))
```

```python
#  5 
for i in range(1,6): 
    print(str(i)+":"+movieTitle[i])
```

### <font color=red>>>> </font>   
###  id=101  10 
###  / 

```python
RecommendMovies= model.recommendProducts(101,10) 
for m in RecommendMovies:
    print(""+ str(m[0]) + \
          ""+ str(movieTitle[m[1]]) + \
          "\n"+ str(m[2]))
    print('='*50)
```

### <font color=red>>>> </font>
###  id=202  10 
###  / 

```python
RecommendUser = model.recommendUsers(202, 10) 
print("id:202 , :{0}id:". \
        format(movieTitle[202]))
print(' ')
for u in RecommendUser:
        print("id {0}   {1}".format( u[0],u[2]))
        print('='*50)
```
